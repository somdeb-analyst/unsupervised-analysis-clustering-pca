# STATISTICAL TEST FOR STRENGTH OF CORRELATIONS (BARTLETT SPHERICITY TEST)

Null Hypothesis: H0: The correlations are not significant.
        
Alternative Hypothesis H1: Correlations are significant.
    
Let us consider the level of significance (alpha) as 0.05 (5%).

```python
from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity
chi_square_value,p_value = calculate_bartlett_sphericity(df_pca_scaled)
p_value
```

Since p-value is less than alpha, we can say that the data gives us strong enough evidence that null hypothesis can be rejected in favour of the alternate hypothesis.

So we can conclude that the correlations among the original variables are strong enough so as to ensure that PCA is applicable.

# TEST FOR ADEQUACY OF SAMPLE SIZE (KMO TEST)

```python
from factor_analyzer.factor_analyzer import calculate_kmo
kmo_all,kmo_model = calculate_kmo(df_pca_scaled)
kmo_model
```

# IMPLEMENTING PCA USING ALL THE 57 ORIGINAL FEATURES

```python
from sklearn.decomposition import PCA
pca = PCA(n_components= len(df_pca_scaled.columns),random_state = 123)
pca_transformed = pca.fit_transform(df_pca_scaled)
```

# OBTAINING THE TRANSFORMED DATAFRAME CONTAINING THE PC SCORES

```python
df_pca_transformed = pd.DataFrame(pca_transformed,columns = ['PC{}'.format(i) for i in range(1,len(df_pca_scaled.columns)+1)])
```

```python
eigen_vectors = pca.components_
print(eigen_vectors)
```

# DECIDING THE OPTIMUM NO OF PRINCIPAL COMPONENTS

# ELBOW METHOD / SCREE PLOT APPROACH

```python
plt.figure(figsize = (20,10))
plt.plot(range(1,58),eigen_values,marker = 'o',markerfacecolor = 'red',markersize = 10)
plt.xticks(range(1,58,2),fontsize = 15)
plt.title('Scree Plot',fontsize = 20)
plt.xlabel('PC number',fontsize = 20)
plt.ylabel('Eigen Value / Explained Variance',fontsize = 20)
plt.grid();
```

Since there is a clear bend / elbow at PC = 7, this method suggests that we can reduce the no of PCs from 57 to 7 without losing too much information / ability to explain the total variance in the dataset

# KAISER RULE APPROACH

This approach tells us to include only the Eigen values > 1. Others are not important as they are not even doing the job of 1 original variable.

```python
plt.figure(figsize = (20,10))
plt.plot(range(1,58),eigen_values,marker = 'o',markerfacecolor = 'red',markersize = 10)
plt.xticks(range(1,58,2),fontsize = 15)
plt.axhline(y = 1,color = 'green',linestyle = '--')
plt.title('Scree Plot',fontsize = 20)
plt.xlabel('PC number',fontsize = 20)
plt.ylabel('Eigen Value / Explained Variance',fontsize = 20)
plt.grid();
```

According to this approach we should select the top 7 PCs (Eigen values > 1) and drop the rest.

# CUMULATIVE EXPLAINED VARIANCE APPROACH

```python
explained_variance_ratio = pca.explained_variance_ratio_
explained_variance_ratio
```

```python
cumulative_explained_variance_ratio = np.cumsum(explained_variance_ratio)
cumulative_explained_variance_ratio
```

```python
plt.figure(figsize = (20,10))
plt.plot(range(1,58),cumulative_explained_variance_ratio,marker = 'o',markerfacecolor = 'red',markersize = 10)
plt.xticks(range(1,58,2),fontsize = 15)
plt.yticks(np.arange(0,1.1,0.1))
plt.axhline(y = 0.9,color = 'green',linestyle = '--')
plt.title('Cumulative Explained Variance',fontsize = 20)
plt.xlabel('PC number',fontsize = 20)
plt.ylabel('Cumulative Explained Variance',fontsize = 20)
plt.grid();
```

So if we want to retain 90% of the information, we have to go with the top 7 PCs and drop the rest.
