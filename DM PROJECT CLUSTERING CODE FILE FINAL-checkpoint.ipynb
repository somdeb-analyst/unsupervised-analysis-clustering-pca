# CHECKING FOR OUTLIERS

```python
#Creating a list of all the numerical columns and displaying the list
num = []
for i in df.columns:
    if df[i].dtype == 'int64' or df[i].dtype == 'float64':
        num.append(i)
num
```

```python
# Checking for outliers by using the boxplot for each numerical feature
from vizad.univariate import plot_univariate_numeric
plot_univariate_numeric ( df,numcols_name = num);
```

There are outliers in most of the features.

# OUTLIER TREATMENT USING THE IQR METHOD

```python
df_new = df.copy()
```

```python
def remove_outliers(col):
    Q1,Q3 = col.quantile([0.25,0.75])
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return lower_bound,upper_bound
```

```python
for i in num:
    lb,ub = remove_outliers(df_new[i])
    df_new[i] = np.where(df_new[i]<lb,lb,df_new[i])
    df_new[i] = np.where(df_new[i]>ub,ub,df_new[i])
```

# CHECKING THE EFFECT OF OUTLIER TREATMENT ON THE NUMERICAL VARIABLES

# CLUSTERING WITH THE DATA WITH OUTLIERS NOT TREATED

# DROPPING  CATEGORICAL FEATURES 

```python
# Creating a list of categorical columns
cat = [col for col in df.columns if df[col].dtype == 'O']
cat
```

```python
# Creating a dataframe with only the numeric columns from the original dataframe
df_num = df.drop(labels = cat,axis = 1)
```

```python
# Checking the top 5 records of the new dataframe consisting of only the numeric columns
df_num.head()
```

# SCALING THE DATA USING ZSCORE TECHNIQUE FROM SKLEARN

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_num_scaled = scaler.fit_transform(df_num)
df_num_scaled = pd.DataFrame(df_num_scaled,columns = df_num.columns)
df_num_scaled.head()
```

# CHECKING THE STATISTICAL SUMMARY OF THE DATA POST SCALING

```python
df_num_scaled.describe().round(2).T
```

All features are now scaled and have a mean of 0 and standard deviation of 1 which is the expected characteristic of data subjected to z-score scaling.

# PERFORMING K-MEANS CLUSTERING ON THE SCALED DATA

# USING ELBOW PLOT / SCREE PLOT APPROACH FOR IDENTIFYING OPTIMUM NUMBER OF CLUSTERS

```python
from sklearn.cluster import KMeans
```

```python
wss = []
k = range(1,11)
for i in k:
    k_means = KMeans (n_clusters = i,random_state = 123)
    k_means.fit(df_num_scaled)
    wss.append(k_means.inertia_)
```

```python
plt.figure(figsize = (10,6))
plt.plot(k,wss,marker = 'o',linestyle = '--')
plt.title('K-Means Clustering')
plt.xlabel('Number of Clusters')
plt.ylabel('WSS')
plt.xticks(k);
```

# IDENTIFYING OPTIMUM NUMBER OF CLUSTERS USING SILHOUETTE SCORE APPROACH

```python
from sklearn.metrics import silhouette_score
```

```python
k = range(2,11)
silhouette_scores = []
for i in k:
    k_means = KMeans (n_clusters = i,random_state = 123)
    k_means.fit(df_num_scaled)
    silhouette_scores.append(silhouette_score (df_num_scaled,k_means.labels_))
```

```python
plt.figure(figsize = (10,6))
plt.plot(k,silhouette_scores,marker = 'o',linestyle = '--')
plt.title('K-Means Clustering')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.xticks(k);
```

From the combined results of the above two approaches we conclude at an optimum number of clusters = 2.

# HIERARCHICAL CLUSTERING (USING DATA WITH OUTLIERS TREATED BY IQR METHOD)

```python
from scipy.cluster.hierarchy import dendrogram,linkage
```

```python
wardlink = linkage(df_num1_scaled,method = 'ward',metric = 'euclidean')
```

```python
plt.figure(figsize = (10,5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Ads')
plt.ylabel('Distance')
dendrogram(wardlink,truncate_mode = 'lastp',p = 5,leaf_rotation = 90.,leaf_font_size = 8.);
```

# FORMING THE CLUSTERS

```python
from scipy.cluster.hierarchy import fcluster
```

```python
clusters = fcluster(wardlink,5,criterion = 'maxclust')
clusters
```
